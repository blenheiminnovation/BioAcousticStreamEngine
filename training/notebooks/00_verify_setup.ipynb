```python
import os
import soundfile as sf

# Quick test: load one ECOSoundSet WAV if already downloaded
audio_dir = "../../datasets/ecosoundset/audio"
wav_files = []
if os.path.isdir(audio_dir):
    for root, _, files in os.walk(audio_dir):
        for f in files:
            if f.endswith(".wav"):
                wav_files.append(os.path.join(root, f))
        if wav_files:
            break

if wav_files:
    test_file = wav_files[0]
    data, sr = sf.read(test_file)
    print(f"Test file:    {os.path.basename(test_file)}")
    print(f"Sample rate:  {sr} Hz")
    print(f"Duration:     {len(data)/sr:.1f} s")
    print(f"Channels:     {1 if data.ndim == 1 else data.shape[1]}")
else:
    print("Dataset not yet downloaded — run Step 6 in the setup guide to fetch ECOSoundSet.")

print("\n✓ Training environment ready")
```
